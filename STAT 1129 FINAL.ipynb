```python
#import random for ai to generate its choice
import random

class rps:
    #setting up the general background and welcome words
    def __init__(self):
        self.playerscore = 0
        self.aiscore = 0
        self.cheat = False
        self.round = 1
        self.gamenum = 1
        self.choicedict = {1:"ROCK",2:"PAPER",3:"SCISSORS"}
        print("Welcome to the game: Rock, Paper, Scissors!\nYou may enter cheat mode by entering 'cheat'")
    
    #set game condition,ex:game points, cheating.
    def condition(self):
        self.numgame = input("How many points you want to play?")
        if self.numgame.isdigit():
            self.numgame = int(self.numgame)
            self.gametowin = int(self.numgame)
            print("You will need to win",self.gametowin,"points to win the game.\n ")
        else:
            if self.numgame == "cheat":
                print("Cheating mode enabled, enjoy your game!")
                self.cheat = True
            elif self.numgame == "nocheat":
                print("Cheating mode disabled, good luck!")
                self.cheat = False
            else:
                print("Please enter a positive integer number for the number of games you want to play")
            self.condition()
    
    #ask for the player to input its choice by enter 1, 2, or 3.
    #check the correctness of input, ask again if incorrect input
    def playermove(self):
        print("GAME",self.gamenum,"ROUND",self.round)
        self.player = input("Please make your choice, 1:Rock 2:Paper 3:Scissors")
        if self.player.isdigit() and (int(self.player) in [1,2,3]):
            self.player_int = int(self.player)
            self.round += 1
        else:
            print("please enter 1 for rock, 2 for paper, and 3 for scissors\n")
            self.playermove()
    
    #randomly generate ai's choice
    def aimove(self):
        self.ai = int(random.randint(1,3))
        if self.cheat == True:
            print("The AI will threw",self.choicedict[self.ai])
    
    #declare the winner, add points
    def judge(self):
        if (self.player_int == 1 and self.ai == 3) or (self.player_int == 2 and self.ai == 1) or (self.player_int == 3 and self.ai == 2):
            print("Player----"+'\033[1;31m%s\033[0m'%self.choicedict[self.player_int]+"--VS--"+'\033[1;31m%s\033[0m'%self.choicedict[self.ai]+"----AI\nYou will receive one point")
            self.playerscore += 1
        elif (self.player_int==self.ai):
            print("Player----"+'\033[1;31m%s\033[0m'%self.choicedict[self.player_int]+"--VS--"+'\033[1;31m%s\033[0m'%self.choicedict[self.ai]+"----AI\nThis is a tie")
        else:
            print("Player----"+'\033[1;31m%s\033[0m'%self.choicedict[self.player_int]+"--VS--"+'\033[1;31m%s\033[0m'%self.choicedict[self.ai]+"----AI\nThe AI will receive one point")
            self.aiscore +=1
        print("---CURRENT SCORE---")
        print('\033[1;34m%s\033[0m'%self.playerscore," Player VS. AI ",'\033[1;34m%s\033[0m'%self.aiscore)
        print("-------------------\n ")
        
    #main function, check if player or ai wins.
    #decide if another round or a winner has spawned.
    #ask the players if they want to play another game.
    def main(self):
        self.condition()
        while self.playerscore != self.gametowin and self.aiscore != self.gametowin:
            self.aimove()
            self.playermove()
            self.judge()
        if self.playerscore == self.gametowin:
            print("You have won the game!!!")
            if self.ai == 0:
                print("")
        elif self.aiscore ==self.gametowin:
            print("You have lost the game~")
        self.onemore = input(" \nDo you want to start another game? enter 'YES' to start, enter anything else to quit")
        if self.onemore == "YES" or self.onemore == "yes":
            self.playerscore = 0
            self.aiscore = 0
            self.round = 1
            self.gamenum += 1
            self.main()
        else:
            print("Thank you for playing this game")
    

if __name__=="__main__":
    round = rps()
    round.main()
```
